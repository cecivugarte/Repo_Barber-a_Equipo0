# VISUALIZACIONES PREVIAS

Considero que el siguiente gráfico puede ser útil para sacar conclusiones, por ende no lo eliminaré.

```python
# Instala Altair si no lo tienes ya instalado
!pip install altair

# Importar las librerías necesarias
import pandas as pd
import altair as alt

# Cargar el archivo de Excel
file_path = '/content/Cecilia_Vidal_database_Excel.xlsx'
excel_data = pd.ExcelFile(file_path)

# Cargar la hoja "Base" y renombrar las columnas
data = excel_data.parse('Base')
data.columns = [
    'Index', 'Comuna', 'Provincia', 'Superficie_KM2', 'Poblacion_Censo2017',
    'Densidad_Poblacional', 'Barberias_2019', 'Barberias_2020', 'Var_Barberias_2019_2020',
    'Barberias_2021', 'Barberias_2022', 'Var_Barberias_2021_2022',
    'Casos_Policiales_2020', 'Var_Casos_2019_2020', 'Casos_Policiales_2021',
    'Var_Casos_2020_2021', 'Casos_Policiales_2022', 'Var_Casos_2021_2022',
    'Casos_Policiales_2023', 'Var_Casos_2022_2023', 'Casos_Policiales_2024',
    'Indice_Criminalidad_2023'
]

# Filtrar columnas necesarias para la visualización
filtered_data = data[['Comuna', 'Densidad_Poblacional', 'Casos_Policiales_2020',
                      'Casos_Policiales_2021', 'Casos_Policiales_2022',
                      'Casos_Policiales_2023', 'Casos_Policiales_2024']]

# Limpiar filas no deseadas y convertir columnas numéricas
filtered_data = filtered_data[filtered_data['Densidad_Poblacional'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
filtered_data['Densidad_Poblacional'] = pd.to_numeric(filtered_data['Densidad_Poblacional'], errors='coerce')
filtered_data['Casos_Policiales_2020'] = pd.to_numeric(filtered_data['Casos_Policiales_2020'], errors='coerce')
filtered_data['Casos_Policiales_2021'] = pd.to_numeric(filtered_data['Casos_Policiales_2021'], errors='coerce')
filtered_data['Casos_Policiales_2022'] = pd.to_numeric(filtered_data['Casos_Policiales_2022'], errors='coerce')
filtered_data['Casos_Policiales_2023'] = pd.to_numeric(filtered_data['Casos_Policiales_2023'], errors='coerce')
filtered_data['Casos_Policiales_2024'] = pd.to_numeric(filtered_data['Casos_Policiales_2024'], errors='coerce')

# Derretir los datos para obtener un formato largo adecuado para Altair
melted_data = filtered_data.melt(
    id_vars=['Comuna', 'Densidad_Poblacional'],
    value_vars=['Casos_Policiales_2020', 'Casos_Policiales_2021',
                'Casos_Policiales_2022', 'Casos_Policiales_2023',
                'Casos_Policiales_2024'],
    var_name='Año',
    value_name='Casos_Policiales'
)

# Ajustar el nombre de los años
melted_data['Año'] = melted_data['Año'].str.extract('(\d+)').astype(int)

# Crear la visualización en Altair
chart = alt.Chart(melted_data).mark_line(point=True).encode(
    x='Año:O',
    y='Casos_Policiales:Q',
    color='Comuna:N',
    tooltip=['Comuna', 'Año', 'Casos_Policiales', 'Densidad_Poblacional']
).properties(
    title='Cambios en Criminalidad en Comunas y Densidad Poblacional'
).interactive()

# Mostrar el gráfico
chart
```

```python
# Guardar la visualización como HTML
chart.save('visualizacionpre01.html')
```

## Previa producuto final

```python

brush = alt.selection_interval(resolve='global')

base = alt.Chart(data).mark_point().encode(
    color=alt.condition(brush, 'Comuna:N', alt.ColorValue('gray')),
    tooltip=['Comuna:N', 'Numero de Barberias 2019:Q', 'Cant Casos Policiales 2019:Q', 'Numero de Barberias 2020:Q', 'Cant Casos Policiales 2020:Q', 'Numero de Barberias 2021:Q', 'Cant Casos Policiales 2021:Q', 'Numero de Barberias 2022:Q', 'Cant Casos Policiales 2022:Q']
).add_selection(
    brush
).properties(
    width=250,
    height=500,
    title='Comuna'
).interactive()

text = base.mark_text(
    align='left',
    baseline='middle',
    dx=7
).encode(
    text='Comuna:N'
)

base.encode(x='Numero de Barberias 2019:Q', y='Cant Casos Policiales 2019:Q') | base.encode(x='Numero de Barberias 2020:Q', y='Cant Casos Policiales 2020:Q') | base.encode(x='Numero de Barberias 2021:Q', y='Cant Casos Policiales 2021:Q') | base.encode(x='Numero de Barberias 2022:Q', y='Cant Casos Policiales 2022:Q')

```
